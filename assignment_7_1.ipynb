9.Which three countries have won the most medals in recent years (from 1984 to 2008)?


```python
coun=odata[odata.Edition_year>=1984]
coun.NOC.value_counts()
```

10.Which country has won the most men's gold medals in singles badminton over the years? Sort the results alphabetically by the player's names.

```python
ba=odata[(odata.Medal=='Gold') & (odata.Discipline=='Badminton') & (odata.Gender=="Men") & (odata.Event=='singles')]
ba.sort_values("Athlete")
```

```python
ba=odata[(odata.Medal=='Gold') & (odata.Discipline=='Badminton') & (odata.Gender=="Men") & (odata.Event=='singles')]
ba.sort_values("Athlete",ascending=True)
```

13.Group by Countries and apply aggregate function min,max,count on over the year
